```python
import numpy as np
import cv2
import os
from random import randint
```

```python
# This script lets you play around with the hsv values to use for the mask

basedir = 'data/with_mask/00000_Mask.jpg'

def nothing(x):
    pass

cv2.namedWindow('Color Capture')

cv2.createTrackbar('M1', 'Color Capture', 0, 255, nothing)
cv2.createTrackbar('S1', 'Color Capture', 0, 255, nothing)
cv2.createTrackbar('V1', 'Color Capture', 0, 255, nothing)
cv2.createTrackbar('M2', 'Color Capture', 0, 255, nothing)
cv2.createTrackbar('S2', 'Color Capture', 0, 255, nothing)
cv2.createTrackbar('V2', 'Color Capture', 0, 255, nothing)

img = cv2.imread(basedir)

while(1):
    cv2.imshow('Color Capture', img)
    
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    
    m1 = cv2.getTrackbarPos('M1', 'Color Capture')
    s1 = cv2.getTrackbarPos('S1', 'Color Capture')
    v1 = cv2.getTrackbarPos('V1', 'Color Capture')

    m2 = cv2.getTrackbarPos('M2', 'Color Capture')
    s2 = cv2.getTrackbarPos('S2', 'Color Capture')
    v2 = cv2.getTrackbarPos('V2', 'Color Capture')
    
    lower_color=np.array([m1, s1, v1])
    upper_color=np.array([m2, s2, v2])
    
    mask = cv2.inRange(hsv, lower_color, upper_color)
    res = cv2.bitwise_and(img, img, mask=mask)
    
    
    cv2.imshow('out', res)
    
cv2.destroyAllWindows()
```

```python
# displays random mask color on 1 image at a time. press any key to cycle to next image, press q to exit

basedir = 'data/with_mask'

for image_name in os.listdir(basedir):
    img = cv2.imread(basedir + '/' + image_name)
    
    # hsv threshold
    lower_color=np.array([89, 15, 117])
    upper_color=np.array([108, 203, 255])
    
    # convert image to hsv
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # color mask
    mask = cv2.inRange(hsv, lower_color, upper_color)
    
    # randomize hue and saturation
    h, s, v = cv2.split(hsv)
    h = (h + randint(0, 255)) % 255
    s = (s - randint(0, 20)) % 255
    shifted = cv2.merge([h, s, v])
    
    # update all pixels where the mask is activated to the shifted color
    hsv[mask == 255] = shifted[mask == 255]
    
    # convert back to BGR
    res = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    
    # display image
    cv2.imshow('hsv1', res)
    
    key = cv2.waitKey(0)
    
    if key & 0xFF == ord('q'):
        break
        
    cv2.destroyAllWindows()
    
cv2.destroyAllWindows()
```
